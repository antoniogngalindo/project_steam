# Extraer data del aquivo steam_games.json.gz

```python
import gzip
import pandas as pd
from io import StringIO
import json
```

- Determinar la ruta del archivo

```python
ruta = 'Datasets/steam_games.json.gz'
```

Primero, creo una lista vacía llamada lineas_json, que utilizaré para almacenar las líneas de texto extraídas del archivo JSON.

Luego, abro el archivo gzip ubicado en la ruta especificada (ruta) en modo de lectura binaria ('rb') - que significa que los datos se leerán y se interpretarán como una secuencia de bytes, en vez de una secuencia de caracteres. 

A continuación, itero sobre cada línea del archivo gzip.

En cada iteración, decodifico la línea del formato binario a una cadena de texto utilizando el formato UTF-8. Luego, agrego la línea decodificada a la lista lineas_json. Hago esto para convertir los datos binarios del archivo gzip a texto legible.

```python
lineas_json = []


with gzip.open(ruta, 'rb') as archivo_gz:
    for linea in archivo_gz:
        lineas_json.append(linea.decode('utf-8')) 
```

Después de leer todas las líneas del archivo, utilizo el método join() para concatenar todas las líneas de texto de la lista lineas_json en una sola cadena llamada contenido_json.

Luego, creo un objeto StringIO llamado archivo_json a partir de la cadena contenido_json. El objeto StringIO proporciona una interfaz de archivo para trabajar con cadenas de texto como si fueran archivos reales.

A continuación, itero sobre cada línea en archivo_json y utilizo la función json.loads() para convertir cada línea en un objeto JSON. Los objetos JSON resultantes se almacenan en una lista llamada datos_json.

```python
contenido_json = ''.join(lineas_json)
archivo_json = StringIO(contenido_json)
datos_json = [json.loads(linea) for linea in archivo_json]
```

Creo un DataFrame de pandas llamado steam_games utilizando la lista de objetos JSON datos_json. Cada objeto JSON se convierte en una fila del DataFrame.

```python
steam_games = pd.DataFrame(datos_json)
steam_games.head()
```

```python
steam_games.tail()
```

Borro todas as lineas que son por completo NaN

```python
steam_games.dropna(how = 'all', inplace= True)
```

Guardo DataFrame en un archivo csv

```python
steam_games.to_csv('steam_games.csv', index=False)
```
