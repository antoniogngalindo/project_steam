# Extraer data del archivo user_reviews.json.gz

```python
import gzip
import pandas as pd
import ast
```

- Determinar ruta del archivo

```python
ruta = 'Datasets/user_reviews.json.gz'
```

- Abro un archivo comprimido en formato gzip y lo leo línea por línea. A medida que recorro cada línea, la decodifico de bytes a caracteres utilizando la codificación 'utf-8' y la evalúo como una expresión de Python utilizando ast.literal_eval(). Luego, agrego los resultados a una lista llamada lineas.

```python
lineas = []

with gzip.open(ruta, 'rb') as archivo:
    for line in archivo:
        line = ast.literal_eval(line.decode('utf-8'))
        lineas.append(line)

```

- Creo el primer DataFrame con la lista

```python
df_user_reviews = pd.DataFrame(lineas)
df_user_reviews.head()
```

- Separo los valores de la lista en la coluna reviews con el metodo .explode()

```python
df_user_reviews = df_user_reviews.explode('reviews')
```

```python
df_user_reviews
```

- Creo un segun DataFrame aplicando el metodo .apply() para generar series con los valores de la coluna reviews

```python
reviews = df_user_reviews['reviews'].apply(pd.Series)
```

```python
reviews.head()
```

- Borro la coluna reviews de la tabla original

```python
df_user_reviews.drop('reviews', axis= 1, inplace= True)
```

- Creo un DataFrame final concatenando la tabla original y la tabla de reviews

```python
user_reviews = pd.concat([df_user_reviews,reviews], axis= 1)
```

```python
user_reviews
```

- se notá que la coluna '0' tiene todos los valores nulos

```python
user_reviews.isna().sum()
```

- Borro la coluna '0'

```python
user_reviews.drop(0, axis = 1, inplace= True)
```

```python
user_reviews
```

- Guardo el DataFrame final en un archivo csv

```python
user_reviews.to_csv('user_reviews.csv', index= False)
```
