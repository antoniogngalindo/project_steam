# Defino la función UserForGenre

En la primera línea, defino la función UserForGenre que toma un parámetro llamado genero.

En la segunda línea, creo una variable llamada filtro que contiene un subconjunto de datos filtrados del DataFrame df_user_games. Utilizo la columna 'most_played_genre' para filtrar los registros que coinciden con el género especificado.

En la tercera línea, verifico si la longitud de filtro es mayor o igual a 1. Esto me permite comprobar si se encontraron registros que coinciden con el género especificado.

Si la longitud de filtro es mayor o igual a 1, significa que se encontraron registros que coinciden con el género especificado.

En la cuarta línea, calculo la suma de las horas jugadas por cada usuario agrupando por su ID. Esto lo guardo en la variable user_playtime.

En la quinta línea, encuentro el usuario con más horas jugadas utilizando el índice con el valor máximo en la columna 'playtime_forever' de user_playtime. Guardo el ID del usuario en la variable most_played_user.

En la sexta línea, creo una lista vacía llamada horas_jugadas para almacenar las horas jugadas por el usuario en cada año.

En la séptima línea, itero sobre los años únicos en el DataFrame filtrado filtro.

En la octava línea, calculo la suma de las horas jugadas por el usuario en el año específico utilizando las condiciones de filtrado. Guardo este valor en la variable playtime.

En la novena línea, verifico si la suma de las horas jugadas es mayor a cero.

En la décima línea, creo un diccionario llamado hora_por_ano que contiene el año como clave y las horas jugadas como valor. Convierto el año y las horas jugadas a enteros antes de agregarlos al diccionario.

En la undécima línea, agrego el diccionario hora_por_ano a la lista horas_jugadas.

En la duodécima línea, retorno una cadena de texto formateada que muestra el usuario con más horas jugadas para el género especificado, junto con la lista de horas jugadas en cada año.

Si la longitud de filtro no es mayor o igual a 1, significa que no se encontraron registros que coincidan con el género especificado. En este caso, retorno un mensaje indicando que el género es incorrecto y se debe intentar nuevamente.

```python
def UserForGenre(genero):
    filtro = df_user_games[df_user_games['most_played_genre'] == genero]
    if len(filtro) >= 1:
        user_playtime = filtro.groupby('user_id')['playtime_forever'].sum().reset_index()
        most_played_user = user_playtime.loc[user_playtime['playtime_forever'].idxmax(), 'user_id']
        
        horas_jugadas = []
        for year in filtro['year'].unique():
            playtime = filtro[(filtro['user_id'] == most_played_user) & (filtro['year'] == year)]['playtime_forever'].sum()
            if playtime > 0:
                hora_por_ano = {'Año': int(year), 'Horas': int(playtime)}
                horas_jugadas.append(hora_por_ano)

        return {f'Usuário com mais horas jogadas para {genero}: {most_played_user} , Horas jogadas: {horas_jugadas}'}
    else:
        return {'No sé encontró el Género'}
```
