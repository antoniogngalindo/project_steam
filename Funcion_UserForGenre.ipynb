```python
import pandas as pd
```

```python
user_items = pd.read_csv('Datasets/users_items.csv')
```

```python
games = pd.read_csv('Datasets/games_genres.csv')
```

```python
user_items.drop(['item_id', 'playtime_2weeks'], axis= 1, inplace= True)
```

```python
df = pd.merge(games, user_items, left_on= 'title', right_on= 'item_name')
```

```python
df
```

```python
df.drop(['title', 'item_name'], axis = 1, inplace= True)
```

```python
df
```

```python
columnas_para_excluir = ['year', 'user_id', 'playtime_forever']
columnas_para_sumar = df.columns.difference(columnas_para_excluir)

df_summed = df.groupby(columnas_para_excluir)[columnas_para_sumar].sum().reset_index()
```

```python
df_summed
```

```python
df_summed['most_played_genre'] = df_summed[columnas_para_sumar].idxmax(axis=1)
```

```python
df_user_games = df_summed[['year', 'user_id', 'playtime_forever', 'most_played_genre']]
```

```python
df_user_games
```

```python
df_user_games = df_user_games[df_user_games['playtime_forever'] != 0]
```

```python
df_user_games[(df_user_games['user_id'] == 'REBAS_AS_F-T') & (df_user_games['most_played_genre'] == 'Adventure')]
```

```python
def UserForGenre(genero):
    filtered_df = df_user_games[df_user_games['most_played_genre'] == genero]
    if len(filtered_df) >= 1:
        user_hours = filtered_df.groupby('user_id')['playtime_forever'].sum().reset_index()
        most_played_user = user_hours.loc[user_hours['playtime_forever'].idxmax(), 'user_id']
        
        horas_jugadas = []
        for year in filtered_df['year'].unique():
            playtime = filtered_df[(filtered_df['user_id'] == most_played_user) & (filtered_df['year'] == year)]['playtime_forever'].sum()
            if playtime > 0:
                hora_por_ano = {'Año': int(year), 'Horas': int(playtime)}
                horas_jugadas.append(hora_por_ano)

        return f'Usuário com mais horas jogadas para {genero}: {most_played_user}, Horas jogadas: {horas_jugadas}'
    else:
        return 'Gênero incorreto. Por favor, tente novamente.'
```

```python
UserForGenre('Action')
```

```python
df_user_games.to_csv('Datasets/UserForGenre.csv', index=False)
```
