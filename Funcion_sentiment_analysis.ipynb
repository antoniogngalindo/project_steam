# Defino la función sentiment_analysis

En la primera línea, defino una función llamada sentiment_analysis que toma un parámetro llamado year (int).

En la segunda línea, creo una variable llamada filtro que contiene un subconjunto de datos filtrados del DataFrame df_sentiment_analysis. Utilizo la columna 'year' para filtrar aquellos registros que coincidan con el año que he especificado.

En la tercera línea, verifico si la longitud de filtro es mayor que 0. Esto me permite comprobar si se encontraron registros que coinciden con el año especificado.

Si la longitud de filtro es mayor que 0, significa que se encontraron registros que coinciden con el año especificado.

En la cuarta línea, selecciono las columnas 'Negativo', 'Neutro' y 'Positivo' del DataFrame filtro y las asigno a la variable sentimientos.

En la quinta línea, utilizo el método to_dict en sentimientos para convertir los datos en un formato de diccionario. Utilizo el parámetro orient='records' para obtener un diccionario que contiene los registros como listas de pares clave-valor.

Si la longitud de filtro no es mayor que 0, significa que no se encontraron registros que coincidan con el año especificado. En este caso, retorno un diccionario que contiene un mensaje indicando que el año no fue encontrado.

```python
def sentiment_analysis( year : int ):
    filtro = df_sentiment_analysis[df_sentiment_analysis['year']== year]
    
    if len(filtro) > 0:
        sentimientos = filtro[['Negativo', 'Neutro', 'Positivo']]

        return sentimientos.to_dict(orient = 'records')

    
    else: 
        return {f'Año no encuentrado!'}
```
