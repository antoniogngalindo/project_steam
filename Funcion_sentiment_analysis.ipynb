```python
years = pd.read_csv('Datasets/games_genres.csv')
```

```python
columnas_keep = ['title', 'year']
years = years[columnas_keep]
```

```python
years
```

```python
sentiment = pd.merge(sentiment, years, left_on='item_name', right_on= 'title')
```

```python
sentiment.drop(['item_name', 'title'], axis = 1, inplace = True)
```

```python
sentiment.dropna(inplace = True)
```

```python
sentiment
```

```python
sentiment.to_csv('Datasets/sentiment_analysis_function.csv', index= False)
```

```python
def sentiment_analysis( año : int ):
    filtro = sentiment[sentiment['year']== año]
    negativo = 0
    neutro = 0
    positivo = 0
    if len(filtro) > 0:
        negativo = filtro[filtro['significado']== 'Negativo']['significado'].count()
        neutro = filtro[filtro['significado']== 'Neutro']['significado'].count()
        positivo = filtro[filtro['significado']== 'Positivo']['significado'].count()
        return {'Negativo': negativo, 'Neutral': neutro, 'Positivo': positivo}
    
    else: 
        return f'Año no encuentrado!'
```

```python
sentiment_analysis(2013)
```

```python
sentiment_analysis(1990)
```
