# Defino la función recomendacion_juego

Defino la funcion recomendacion_juego que recibi un parametro id_producto que es de tipo inteiro.

Si el ID del producto está en los valores de la columna item_id del DataFrame df_recomendacion_juego, entonces verifico si el ID del producto proporcionado está presente en la columna item_id del DataFrame df_recomendacion_juego.

Obtengo el índice del producto en el DataFrame df_recomendacion_juego y lo guardo en la variable index.

Utilizo la variable similitud que guarda mi matriz de similitud.

Ordeno las similitudes en orden descendente y obtengo los índices de los productos más similares al producto de interés (excluyendo el propio producto) y los almaceno en la variable similar_indices.

Extraigo los registros de los productos similares del DataFrame df_recomendacion_juego utilizando los índices almacenados en similar_indices. Selecciono las columnas item_id y 'title de esos registros y los guardo en la variable similar_products.

Devuelvo los productos similares en formato de diccionario:

Si el ID del producto no se encuentra en el DataFrame, devuelvo un mensaje indicando que el ID no fue encontrado.

```python
def recomendacion_juego(id_producto: int):
    
    if id_producto in df_recomendacion_juego['item_id'].values:
        index = df_recomendacion_juego[df_recomendacion_juego['item_id'] == id_producto].index[0]
        similarities = similitud
        similar_indices = similarities[index].argsort()[::-1][1:6]
        similar_products = df_recomendacion_juego.iloc[similar_indices][['item_id', 'title']]
        return similar_products.to_dict(orient='records')
    else:
        return 'ID no encontrado'
```
