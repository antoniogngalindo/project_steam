# Defino la función UsersRecommend

En la primera línea, defino la función UsersRecommend que toma un parámetro año de tipo entero.

En la segunda línea, creo una variable llamada filtro que contiene un subconjunto de datos filtrados del DataFrame reviews_gb. Se seleccionan los registros donde el valor de la columna 'year' coincide con el año especificado.

En la tercera línea, verifico si la longitud de filtro es mayor a 1. Esto me permite comprobar si se encontraron registros para el año especificado.

Si la longitud de filtro es mayor a 1, significa que se encontraron registros para ese año.

En la cuarta línea, creo una variable llamada recomend que contiene un subconjunto de datos donde se filtran solo las recomendaciones positivas (valor True) del DataFrame filtro.

En la quinta línea, selecciono las tres mejores recomendaciones en base a la columna 'polaridad' utilizando el método nlargest del DataFrame recomend. Estas recomendaciones se guardan en la variable polaridad. Además, se reinicia el índice del DataFrame utilizando el parámetro drop=True.

En la sexta línea, creo un diccionario vacío llamado resultado para almacenar los resultados de las mejores recomendaciones.

En la séptima línea, itero sobre los índices y las filas del DataFrame polaridad utilizando iterrows(). El índice i y la fila row representan cada una de las mejores recomendaciones.

En la octava línea, obtengo el título del juego de cada fila y lo guardo en la variable game.

En la novena línea, creo una clave para el diccionario resultado que representa el puesto de la recomendación utilizando el índice i y el índice inicial de polaridad. Asigno el título del juego como valor.

En la décima línea, retorno el diccionario resultado que contiene las mejores recomendaciones de juegos para el año especificado.

Si la longitud de filtro no es mayor a 1, significa que no se encontraron registros para el año especificado. En este caso, retorno un diccionario con un mensaje indicando que el año no fue encontrado.

```python
def UsersRecommend(año: int):
    filtro = reviews_gb[reviews_gb['year'] == año]
    if len(filtro) > 1:
        recomend = filtro[filtro['recommend'] == True]
        polaridad = recomend.nlargest(3, 'polaridad').reset_index(drop= True)
        resultado = {}
        for i, row in polaridad.iterrows():
            game = "Puesto " + str(i + 1 - polaridad.index[0])
            resultado[game] = row['title']
        
        return resultado
    else:
        return {f'Año no encuentrado!'}
        
    
```

```python
UsersRecommend(1999)
```

```python
UsersRecommend(2019)
```

# Defino la función UsersNotRecommend

En la primera línea, defino la función UsersNotRecommend que toma un parámetro año de tipo entero.

En la segunda línea, creo una variable llamada filtro que contiene un subconjunto de datos filtrados del DataFrame reviews_gb. Se seleccionan los registros donde el valor de la columna 'year' coincide con el año especificado.

En la tercera línea, verifico si la longitud de filtro es mayor a 1. Esto me permite comprobar si se encontraron registros para el año especificado.

Si la longitud de filtro es mayor a 1, significa que se encontraron registros para ese año.

En la cuarta línea, creo una variable llamada recomend que contiene un subconjunto de datos donde se filtran solo las recomendaciones negativas (valor False) del DataFrame filtro.

En la quinta línea, selecciono las tres peores recomendaciones en base a la columna 'polaridad' utilizando el método nsmallest del DataFrame recomend. Estas recomendaciones se guardan en la variable polaridad. Además, se reinicia el índice del DataFrame utilizando el parámetro drop=True.

En la sexta línea, creo un diccionario vacío llamado resultado para almacenar los resultados de las peores recomendaciones.

En la séptima línea, itero sobre los índices y las filas del DataFrame polaridad utilizando iterrows(). El índice i y la fila row representan cada una de las peores recomendaciones.

En la octava línea, obtengo el título del juego de cada fila y lo guardo en la variable game.

En la novena línea, creo una clave para el diccionario resultado que representa el puesto de la recomendación utilizando el índice i y el índice inicial de polaridad. Asigno el título del juego como valor.

En la décima línea, retorno el diccionario resultado que contiene las peores recomendaciones de juegos para el año especificado.

Si la longitud de filtro no es mayor a 1, significa que no se encontraron registros para el año especificado. En este caso, retorno un diccionario con un mensaje indicando que el año no fue encontrado.

```python
def UsersNotRecommend( año : int ):
    filtro = reviews_gb[reviews_gb['year'] == año]
    if len(filtro) > 1:
        recomend = filtro[filtro['recommend'] == False]
        polaridad = recomend.nsmallest(3, 'polaridad').reset_index(drop= True)
        resultado = {}
        for i, row in polaridad.iterrows():
            game = "Puesto " + str(i + 1 - polaridad.index[0])
            resultado[game] = row['title']
        
        return resultado
    else:
        return {f'Año no encuentrado!'}
```
