Calculo los outliers de la columna 'playtime_forever' y los elimino.

```python
Q1 = games['playtime_forever'].quantile(0.25)
Q3 = games['playtime_forever'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = games[(games['playtime_forever'] < lower_bound) | (games['playtime_forever'] > upper_bound)]


outliers
```

```python
games = games.drop(outliers.index)
```

```python
games.drop_duplicates(inplace=True)
```

# Defino la funciÃ³n UsersRecommend

En la primera lÃ­nea, defino la funciÃ³n UsersRecommend que toma un parÃ¡metro aÃ±o de tipo entero.

En la segunda lÃ­nea, creo una variable llamada filtro que contiene un subconjunto de datos filtrados del DataFrame reviews_gb. Se seleccionan los registros donde el valor de la columna 'year' coincide con el aÃ±o especificado.

En la tercera lÃ­nea, verifico si la longitud de filtro es mayor a 1. Esto me permite comprobar si se encontraron registros para el aÃ±o especificado.

Si la longitud de filtro es mayor a 1, significa que se encontraron registros para ese aÃ±o.

En la cuarta lÃ­nea, creo una variable llamada recomend que contiene un subconjunto de datos donde se filtran solo las recomendaciones positivas (valor True) del DataFrame filtro.

En la quinta lÃ­nea, selecciono las tres mejores recomendaciones en base a la columna 'polaridad' utilizando el mÃ©todo nlargest del DataFrame recomend. Estas recomendaciones se guardan en la variable polaridad. AdemÃ¡s, se reinicia el Ã­ndice del DataFrame utilizando el parÃ¡metro drop=True.

En la sexta lÃ­nea, creo un diccionario vacÃ­o llamado resultado para almacenar los resultados de las mejores recomendaciones.

En la sÃ©ptima lÃ­nea, itero sobre los Ã­ndices y las filas del DataFrame polaridad utilizando iterrows(). El Ã­ndice i y la fila row representan cada una de las mejores recomendaciones.

En la octava lÃ­nea, obtengo el tÃ­tulo del juego de cada fila y lo guardo en la variable game.

En la novena lÃ­nea, creo una clave para el diccionario resultado que representa el puesto de la recomendaciÃ³n utilizando el Ã­ndice i y el Ã­ndice inicial de polaridad. Asigno el tÃ­tulo del juego como valor.

En la dÃ©cima lÃ­nea, retorno el diccionario resultado que contiene las mejores recomendaciones de juegos para el aÃ±o especificado.

Si la longitud de filtro no es mayor a 1, significa que no se encontraron registros para el aÃ±o especificado. En este caso, retorno un diccionario con un mensaje indicando que el aÃ±o no fue encontrado.

```python
def UsersRecommend(aÃ±o: int):
    filtro = reviews_gb[reviews_gb['year'] == aÃ±o]
    if len(filtro) > 1:
        recomend = filtro[filtro['recommend'] == True]
        polaridad = recomend.nlargest(3, 'polaridad').reset_index(drop= True)
        resultado = {}
        for i, row in polaridad.iterrows():
            game = "Puesto " + str(i + 1 - polaridad.index[0])
            resultado[game] = row['title']
        
        return resultado
    else:
        return {f'AÃ±o no encuentrado!'}
        
    
```
